#### Imports

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

#### url in a variable

```python
website = 'https://scholar.google.co.id/citations?user=KGH0e8cAAAAJ&hl=en&cstart=0&pagesize=100'
```

#### Get URL

```python
response = requests.get(website)
```

#### Soup Object

```python
soup = BeautifulSoup(response.content, 'html.parser')
```

#### Results

```python
results = soup.find_all('tr', {'class':'gsc_a_tr'})
```

### Get data single page

```python
title = []
author = []
journal = []
cited = []
year = []

for result in results:
    
    # title
    try:
        title.append(result.find('a', {'class':'gsc_a_at'}).get_text())
    except:
        title.append('n/a')
    
    # author
    try:
        author.append(result.findAll('div', {'class':'gs_gray'})[0].get_text())
    except:
        author.append('n/a')
        
    # journal
    try:
        journal.append(result.findAll('div', {'class':'gs_gray'})[1].get_text())
    except:
        journal.append('n/a')
        
    # cited
    try:
        cited.append(result.find('td', {'class':'gsc_a_c'}).get_text())
    except:
        cited.append('n/a')
    
    # year
    try:
        year.append(result.find('td', {'class':'gsc_a_y'}).get_text())
    except:
        year.append('n/a')
```

```python
title
```

#### Dataframe

```python
scholar = pd.DataFrame({'Title':title, 'Author':author, 'Journal':journal, 'Cited':cited, 'Year':year})
```

```python
scholar
```

#### Data Cleaning

```python
scholar['Journal'] = scholar['Journal'].apply(lambda x: x.replace('\xa0','').split(',')[0:-1])
```

```python
def journalStr(x):
    return (','.join(x))
```

```python
scholar['Journal'] = scholar['Journal'].apply(journalStr)
```

# Get all data

```python
link1 = 'https://scholar.google.co.id/citations?user=KGH0e8cAAAAJ&hl=en&cstart=0&pagesize=100'
link2 = 'https://scholar.google.co.id/citations?user=KGH0e8cAAAAJ&hl=en&cstart=100&pagesize=100'
site = [link1, link2]

title = []
author = []
journal = []
cited = []
year = []

for sequence in site:
    
    # site in a variable
    response = requests.get(sequence)
    
    # soup object
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # results
    results = soup.find_all('tr', {'class':'gsc_a_tr'})
    
    for result in results:
    
        # title
        try:
            title.append(result.find('a', {'class':'gsc_a_at'}).get_text())
        except:
            title.append('n/a')

        # author
        try:
            author.append(result.findAll('div', {'class':'gs_gray'})[0].get_text())
        except:
            author.append('n/a')

        # journal
        try:
            journal.append(result.findAll('div', {'class':'gs_gray'})[1].get_text())
        except:
            journal.append('n/a')

        # cited
        try:
            cited.append(result.find('td', {'class':'gsc_a_c'}).get_text())
        except:
            cited.append('n/a')

        # year
        try:
            year.append(result.find('td', {'class':'gsc_a_y'}).get_text())
        except:
            year.append('n/a')
```

```python
scholar = pd.DataFrame({'Title':title, 'Author':author, 'Journal':journal, 'Cited':cited, 'Year':year})
```

```python
scholar
```

#### Data Cleaning

```python
scholar['Journal'] = scholar['Journal'].apply(lambda x: x.replace('\xa0','').split(',')[0:-1])
```

```python
def journalStr(x):
    return (','.join(x))
```

```python
scholar['Journal'] = scholar['Journal'].apply(journalStr)
```
